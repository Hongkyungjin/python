# 연령별 인구 데이터 살펴보기 (pandas)

## 2019년 2월 기준 연령별 인구 데이터

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('age.csv', encoding='cp949')
df.columns = df.columns.str.replace('2019년02월_계_','')
df['행정구역'] = df['행정구역'].str.split('(', n=1, expand=True)
df['행정구역'] = df['행정구역'].str.strip()
df.head()
```

## 광역자치단체 자료만 불러오기

```python
filter = ['서울특별시', '인천광역시', '부산광역시', '대구광역시', '대전광역시', '광주광역시', \
            '울산광역시', '경기도', '강원도', '충청북도', '충청남도', '전라북도', \
          '전라남도', '경상북도', '경상남도', '제주특별자치도']
df2 = df.loc[df['행정구역'].isin(filter), :]
df2['행정구역'] = ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', \
            'Ulsan', 'Gyeonggi', 'Gangwon', 'Chungbuk', 'Chungnam', \
               'Jeonbuk', 'Jeonnam', 'Gyeongbuk', 'Gyeongnam', 'Jeju']
df2.head()
```

```python
df2.tail()
```

## 광역자치단체 연령별 인구 표 만들기

```python
plt.style.use('ggplot')
province_list = ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', \
            'Ulsan', 'Gyeonggi', 'Gangwon', 'Chungbuk', 'Chungnam', \
                 'Jeonbuk', 'Jeonnam', 'Gyeongbuk', 'Gyeongnam', 'Jeju']
results = {province_list[i] : [int(str(value).replace(',','')) for \
                               value in df2.iloc[i, 3:]] for i in range(len(df2.index))}
df3 = pd.DataFrame(results)
df3.head()
```

```python
df3.tail()
```

## 특별시, 광역시 시각화하기

```python
df4 = df3.iloc[:, :7]
df4.plot(figsize=(15, 10))
plt.title('Population Distribution by Age, 2019', fontsize = 25)
plt.savefig('population1.png', dpi=400, bbox_inches='tight')
plt.show()
```

## 도 시각화하기

```python
df5 = df3.iloc[:, 7:]
df5.plot(figsize=(15, 10))
plt.title('Population Distribution by Age, 2019', fontsize = 25)
plt.savefig('population2.png', dpi=400, bbox_inches='tight')
plt.show()
```

# 성별 인구 데이터 살펴보기 (pandas)

## CSV 파일에서 추출하기

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

dfa = pd.read_csv('gender.csv', encoding='cp949')
dfa.columns = dfa.columns.str.replace('2019년02월_','')
dfa['행정구역'] = dfa['행정구역'].str.split('(', n=1, expand=True)
dfa['행정구역'] = dfa['행정구역'].str.strip()
dfa.head()
```

## 성별 데이터로 나누기

```python
dfm = dfa.iloc[:, 3:104]
dfm = pd.concat([dfa['행정구역'], dfm], axis=1)
dfm.head()
```

```python
dff = dfa.iloc[:, 106:]
dff = pd.concat([dfa['행정구역'], dff], axis=1)
dff.head()
```

## 항아리 모양으로 시각화하기

```python
m = [-int(str(value).replace(',','')) for value in dfm.iloc[0, 1:]] 
f = [int(str(value).replace(',','')) for value in dff.iloc[0, 1:]] 
plt.title('Gender Distribution, 2019 Seoul', fontsize = 20)
plt.barh(range(101), m, label='Male')
plt.barh(range(101), f, label='Female')
plt.legend()
plt.savefig('population5.png', dpi=400, bbox_inches='tight')
plt.show()
```

## 광역자치단체 성별 인구 표 만들기

```python
def comparison(male, female):
    if male > female:
        return 'More Male'
    else:
        return 'More Female'

filter = ['서울특별시', '인천광역시', '부산광역시', \
          '대구광역시', '대전광역시', '광주광역시', \
            '울산광역시', '경기도', '강원도', '충청북도', '충청남도', \
          '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도']
dfc = dfa.loc[dfa['행정구역'].isin(filter), :]
dfc = dfc.iloc[:, [0, 1, 104]]
dfc.columns = ['Region', 'Male', 'Female']
dfc['Region'] = ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', \
            'Ulsan', 'Gyeonggi', 'Gangwon', 'Chungbuk', 'Chungnam', 'Jeonbuk', \
                 'Jeonnam', 'Gyeongbuk', 'Gyeongnam', 'Jeju']
dfc['Comparison'] = dfc.apply (lambda row: comparison(row.Male, row.Female), axis=1)
dfc.head()
```

```python
dfc.tail()
```

## 제주도는 아직도 삼다도(三多島)일까?

### 파이 그래프로 나타내기

```python
dfj = [int(str(value).replace(',','')) for value in \
       dfc[dfc['Region'].str.startswith('Jeju')].loc[:, \
                                                     ['Male', 'Female']].values.tolist()[0]]
sej = pd.Series(dfj, index=['Male', 'Female'])
sej.plot.pie(autopct='%1.1f%%', shadow=True, startangle=90, \
             fontsize = 20, figsize=(15, 10))
plt.title('Population Distribution by Gender, 2019 Jeju with ' + \
          str(sej.sum()) + ' People', fontsize = 20)
plt.axis('equal')
plt.tight_layout()
plt.savefig('population6.png', dpi=400, bbox_inches='tight')
plt.show()
```

### 산점도로 나타내기

```python
m = [int(str(value).replace(',','')) for value in dfm.iloc[3795, 1:]] 
f = [int(str(value).replace(',','')) for value in dff.iloc[3795, 1:]] 
plt.title('Gender Distribution, 2019 Jeju', fontsize = 20)
plt.scatter(m, f, c= range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel('Male')
plt.ylabel('Female')
plt.savefig('population7.png', dpi=400, bbox_inches='tight')
plt.show()
```
